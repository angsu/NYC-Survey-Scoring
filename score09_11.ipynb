```python
nvarblue = pd.DataFrame(index=np.arange(100), columns = variable)
nvarorange = pd.DataFrame(index=np.arange(100), columns = variable)
nvaryellow = pd.DataFrame(index=np.arange(100), columns = variable)
nvarblue['question'] = 'NaN'
nvarorange['question'] = 'NaN'
nvaryellow['question'] = 'NaN'
nvarblue['item'] = 'NaN'
nvarorange['item'] = 'NaN'
nvaryellow['item'] = 'NaN'

nvarcolor = [nvarblue,nvarorange,nvaryellow]
scolor = ['blue','orange','yellow']

#get list of all variables s_q2a_1, s_q2a_2, etc...
for i in range(0,len(color)):
    for j in range(0, len(variable)):
        for k in range(0, len(color[i][variable[j]])):
            if pd.isnull(color[i][variable[j]][k]):
                continue
            else:
                l = 4*k
                nvarcolor[i][variable[j]][l] = color[i][variable[j]][k] + '_1'
                nvarcolor[i]['question'][l] = color[i]['item'][k]
                nvarcolor[i]['item'][l] = scolor[i] + str(int(color[i]['Unnamed: 0'][k]))
                nvarcolor[i][variable[j]][l+1] = color[i][variable[j]][k] + '_2'
                nvarcolor[i][variable[j]][l+2] = color[i][variable[j]][k] + '_3'
                nvarcolor[i][variable[j]][l+3] = color[i][variable[j]][k] + '_4'
            
```

```python
first2009 = all[yryr[0]].iloc[7::4, :]
first2010 = all[yryr[1]].iloc[7::4, :]
first2011 = all[yryr[2]].iloc[2::4, :]
firstyr = [first2009,first2010,first2011]

negativeyr = ['neg2009','neg2010','neg2011']

#scores the questions
for j in range(0,len(all)):
    for i in range(0,len(all[yryr[j]])):
        #if not pd.isnull(all[yryr[j]]['variable'][i]):
        if all[yryr[j]]['variable'][i] in list(firstyr[j]['variable']):
            if all[yryr[j]]['variable'][i] in list(negtype[negativeyr[j]]):
                n = all[yryr[j]].shape[1]
                line = -2 * all[yryr[j]].iloc[i,1:n-1] - 1 * all[yryr[j]].iloc[i+1,1:n-1] + all[yryr[j]].iloc[i+2,1:n-1] + 2 * all[yryr[j]].iloc[i+3,1:n-1]                     
            else:              
                n = all[yryr[j]].shape[1]
                line = 2 * all[yryr[j]].iloc[i,1:n-1] + all[yryr[j]].iloc[i+1,1:n-1] - 1 *all[yryr[j]].iloc[i+2,1:n-1] - 2 * all[yryr[j]].iloc[i+3,1:n-1]                     
            #gets sum of the first four responses; if only four, then will be 100; if five, may be less than 100
            sum = all[yryr[j]].iloc[i,1:n-1] + all[yryr[j]].iloc[i+1,1:n-1] + all[yryr[j]].iloc[i+2,1:n-1] + all[yryr[j]].iloc[i+3,1:n-1]
            sum.replace(0, np.nan, inplace=True)
            line = line / sum
            all[yryr[j]].iloc[i,1:n-1] = line
    
```
