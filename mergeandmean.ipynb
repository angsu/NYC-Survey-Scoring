```python
import pandas as pd
import numpy as np

from openpyxl import Workbook, load_workbook
from openpyxl.utils.dataframe import dataframe_to_rows
```

```python
nine = pd.read_excel('scores091011.xlsx', sheet_name='Sheet1')
twelve = pd.read_excel('scores12_18.xlsx', sheet_name='Sheet1')

nine=nine.iloc[1:]
nine.set_index('Unnamed: 0',inplace=True)
names9 = nine.columns.values
names9 = pd.Series(names9)
nine.columns = names9.str.split('.').str[0]



twelve = twelve.iloc[1:]
twelve.set_index('Unnamed: 0',inplace=True)
df = twelve['school_name']
twelve = twelve.drop('school_name',axis=1)
names12 = twelve.columns.values
names12 = pd.Series(names12)
twelve.columns = names12.str.split('.').str[0]
```

```python
pd.concat([nine,twelve],axis=1,sort=False)
```

```python
df
```

```python
#get mean
result = pd.DataFrame()
result = pd.concat([nine,twelve],axis=1,sort=False)
for j in range(3,len(result)):
    if pd.isnull(result.iloc[j,0]):
        idx = result.index[j]
        result.iloc[j,0] = df.loc[idx]
result['sch_mean_allyrs'] = 'NaN'
result = pd.concat([result[['sch_mean_allyrs']], result.drop('sch_mean_allyrs', axis=1)], axis=1)
result = result.reset_index()

for i in range(3,len(result)):
    formean = result.iloc[i,4:]
    result.loc[i,'sch_mean_allyrs'] = formean.mean(skipna = True)
    
```

```python
result
```

```python
result1 = result
sch_meanyr = ['sch_mean2009','sch_mean2010','sch_mean2011','sch_mean2012','sch_mean2013','sch_mean2014','sch_mean2015','sch_mean2016','sch_mean2017','sch_mean2018']
for i in range(0, len(sch_meanyr)):
    idx = len(sch_meanyr)-1-i
    result1[sch_meanyr[idx]]='NaN'
    result1 = pd.concat([result1[[sch_meanyr[idx]]], result1.drop(sch_meanyr[idx], axis=1)], axis=1)
    result1 = result1.reset_index(drop = True)

yr = [2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
for i in range(3,len(result1)):
    for j in range(0,len(sch_meanyr)):
        s = result1.iloc[1]        
        a = list(s.index.values[(s == yr[j])])
        formean = result1.iloc[i][a]
        result1.loc[i,sch_meanyr[j]] = formean.mean(skipna = True)
        

result1
```

```python
result1 = pd.concat([result1[['sch_mean_allyrs']], result1.drop('sch_mean_allyrs', axis=1)], axis=1)
result1 = pd.concat([result1[['sch_type']], result1.drop('sch_type', axis=1)], axis=1)
result1 = pd.concat([result1[['school_name']], result1.drop('school_name', axis=1)], axis=1)
result1 = pd.concat([result1[['index']], result1.drop('index', axis=1)], axis=1)
result1
```

```python
wb = load_workbook('finalscores.xlsx')
sheet1 = wb.create_sheet('Sheet1',0)
active = wb['Sheet1']
# Write dataframe to active worksheet
for x in dataframe_to_rows(result1):
    active.append(x)
    
# Save workbook to write
wb.save('finalscores.xlsx')
```
