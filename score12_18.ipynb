```python
#lists of variables for looping
files_q = ['q2012.csv','q2013.csv','q2014.csv','q2015.csv','q2016.csv','q2017.csv','q2018.csv']
yr = ['2012','2013','2014','2015','2016','2017','2018']
yryr = ['yr2012','yr2013','yr2014','yr2015','yr2016','yr2017','yr2018']
queyr = ['que2012','que2013','que2014','que2015','que2016','que2017','que2018']
qtypeyr = ['qtype2012','qtype2013','qtype2014','qtype2015','qtype2016','qtype2017','qtype2018']
xyr = ['x2012','x2013','x2014','x2015','x2016','x2017','x2018']
```

```python
qtype = pd.DataFrame()

#create file qtype containing questions and variables for all years
for idx, fil in enumerate(files_q):
    data = pd.read_csv(files_q[idx], engine='python')
    data.columns = ['ind','que','x']
    if fil =='q2012.csv':
        data = data.iloc[4:]
    if fil =='q2013.csv':
        data = data.iloc[4:]
    else:
        data = data.iloc[3:]
    #n = names[idx]
    data = data.reset_index(drop=True)
    data = data.rename(columns=lambda x : x + yr[idx])
    data.insert(3, "qtype" + yr[idx], 0, True)
    #qtype = qtype.append(data)
    qtype = pd.concat([qtype, data], axis=1, sort=False)
```

```python
#get the # of response options for each question (4, 5, ..)

#updates qtype dataframe with type of question (4 or 5 response question)
for idx,col in enumerate(queyr):
    t = 'qtype' + yr[idx]
    for i in range(0,len(qtype)):
        if pd.isnull(qtype[xyr[idx]][i]):
            qtype[t][i-last]=last
            break
        if not pd.isnull(qtype[col][i]):
            try:
                if not pd.isnull(qtype[col][i+1]):
                    qtype[t][i] = 1
                    last = 1
                elif not pd.isnull(qtype[col][i+2]):
                    qtype[t][i] = 2
                    last = 2
                elif not pd.isnull(qtype[col][i+3]):
                    qtype[t][i] = 3
                    last = 3
                elif not pd.isnull(qtype[col][i+4]):
                    qtype[t][i] = 4
                    last = 4
                elif not pd.isnull(qtype[col][i+5]):
                    qtype[t][i] = 5
                    last = 5
                elif not pd.isnull(qtype[col][i+6]):
                    qtype[t][i] = 6
                    last = 6
            except(KeyError): continue
        else:continue
```

```python
#merge colored files with qtype on relevant questions
def mergeqt(color):
    result=color
    for i in range(0,len(qtypeyr)):
        typeyr = qtype[[xyr[i],qtypeyr[i],queyr[i]]]
        result[xyr[i]] = result[xyr[i]].fillna(0)
        typeyr[xyr[i]]=typeyr[xyr[i]].fillna(1)
        result = pd.merge(result, typeyr, how='left',on=xyr[i])
    return result
    
```

```python
orangewtype = mergeqt(orange)
bluewtype = mergeqt(blue)
yellowwtype = mergeqt(yellow)
```

```python
#add rows for multiple response questions

def addresp(merge):   
    for k in range(0,len(frames_t)):
        for i in range(0,len(merge[yryr[k]])):
            try:
                idx = frames_t[k].index[frames_t[k]['index']==merge[yryr[k]][xyr[k]][i]][0]
                #append rest of the responses obtained from # in qtype

                for j in range(1,int(merge[yryr[k]][qtypeyr[k]][i])):
                    line = pd.DataFrame(frames_t[k].iloc[int(idx + j)]).T
                    line.index=[i + j / merge[yryr[k]][qtypeyr[k]][i]]
                    merge[yryr[k]] = merge[yryr[k]].append(line, ignore_index=False)
            except(IndexError): continue
    return
```

```python
#scores the questions

negativeyr = ['neg2012','neg2013','neg2014','neg2015','neg2016','neg2017','neg2018']

for j in range(0,len(all)):
    for i in range(0,len(all[yryr[j]])):
        if not pd.isnull(all[yryr[j]][xyr[j]][i]):
            if all[yryr[j]][xyr[j]][i] in list(negtype[negativeyr[j]]):
                n = all[yryr[j]].shape[1]
                line = -2 * all[yryr[j]].iloc[i,2:n-2] - 1 * all[yryr[j]].iloc[i+1,2:n-2] + all[yryr[j]].iloc[i+2,2:n-2] + 2 * all[yryr[j]].iloc[i+3,2:n-2]                     
            else:
                n = all[yryr[j]].shape[1]
                line = 2 * all[yryr[j]].iloc[i,2:n-2] + all[yryr[j]].iloc[i+1,2:n-2] - 1 *all[yryr[j]].iloc[i+2,2:n-2] - 2 * all[yryr[j]].iloc[i+3,2:n-2]
            
            sum = all[yryr[j]].iloc[i,2:n-2] + all[yryr[j]].iloc[i+1,2:n-2] + all[yryr[j]].iloc[i+2,2:n-2] + all[yryr[j]].iloc[i+3,2:n-2]
            sum.replace(0, np.nan, inplace=True)
            line = line / sum
            all[yryr[j]].iloc[i,2:n-2] = line
    
```
